## Chương 6: Morphological Image Processing

#### __Bài tập 1:__ Hiển thị bằng một hình ảnh nhị phân rằng phép mở và phép đóng hình thái học là các phép toán đối ngẫu. (Gợi ý: áp dụng phép mở trên tiền cảnh của hình ảnh và phép đóng trên nền của hình ảnh với cùng một phần tử cấu trúc)

#### __Giải thích code__

#### 1. Nhập các thư viện cần thiết

```python
# binary_opening và binary_closing từ skimage.morphology để thực hiện phép mở và phép đóng hình thái học
# disk từ skimage.morphology để tạo phần tử cấu trúc hình đĩa
from skimage.morphology import binary_opening, binary_closing, disk
# invert từ skimage.util để đảo ngược ảnh
from skimage.util import invert                 
# rgb2gray từ skimage.color để chuyển đổi ảnh sang grayscale
from skimage.color import rgb2gray              
# imread từ skimage.io để đọc ảnh
from skimage.io import imread, imsave           
# pylab và matplotlib.pyplot để vẽ đồ thị và hiển thị ảnh
import pylab                                    
import matplotlib.pyplot as plt
```

#### 2. Đọc và chuyển đổi ảnh sang nhị phân

```python
# Đọc ảnh từ đường dẫn và chuyển đổi ảnh sang grayscale
im = rgb2gray(imread('../images/circles.jpg'))  
# Chuyển đổi ảnh sang grayscale thành ảnh nhị phân bằng cách đặt các giá trị pixel nhỏ hơn hoặc bằng 0.5 thành 0 và các giá trị lớn hơn 0.5 thành 1
im[im <= 0.5] = 0
im[im > 0.5] = 1
# Sử dụng pylab.gray() để hiển thị ảnh ở chế độ grayscale
pylab.gray()
```

#### 3. Định nghĩa hàm plot_image để hiển thị ảnh với tiêu đề

```python
# Hàm plot_image nhận vào một ảnh image và một tiêu đề title
def plot_image(image, title):			
	# Sử dụng plt.imshow để hiển thị ảnh
	plt.imshow(image)					
	# Sử dụng plt.title để đặt tiêu đề cho ảnh
	plt.title(title)					
	# Sử dụng plt.axis('off') để tắt các trục tọa độ
	plt.axis('off')						
```

#### 4. Áp dụng phép mở và phép đóng hình thái học và hiển thị kết quả

```python
# Tạo một hình ảnh với kích thước 20x10 inch
pylab.figure(figsize=(20,10))                       
# Hiển thị ảnh gốc im trong ô đầu tiên (1, 3, 1)
pylab.subplot(1,3,1), plot_image(im, 'original')    
# Áp dụng phép mở hình thái học với phần tử cấu trúc hình đĩa có kích thước 12 và lưu kết quả vào im1. Hiển thị ảnh đã mở trong ô thứ hai (1, 3, 2)
im1 = binary_opening(im, disk(12))
pylab.subplot(1,3,2), plot_image(im1, 'opening with disk size ' + str(12))
# Lưu ảnh đã mở thành file png
imsave('opening.jpg', im1)
# Áp dụng phép đóng hình thái học với phần tử cấu trúc hình đĩa có kích thước 6 và lưu kết quả vào im1. Hiển thị ảnh đã đóng trong ô thứ ba (1, 3, 3)
im1 = binary_closing(im, disk(6))
pylab.subplot(1,3,3), plot_image(im1, 'closing with disk size ' + str(6))
# Lưu ảnh đã đóng thành file png
imsave('closing.png', im1)
# Hiển thị hình ảnh
pylab.show()
```
