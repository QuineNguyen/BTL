#### __Giải thích code__

#### 1. Nhập các thư viện cần thiết

```python
import numpy as np                  
from scipy import ndimage              
import imageio.v2 as imageio        
import matplotlib.pyplot as plt     
```

#### 2. Các thao tác với ảnh

```python
# Đọc ảnh: 
image = imageio.imread('../images/goldengate.jpg')

# Phóng to ảnh (zoom): Phóng to gấp 2 lần trên hai chiều không gian
zoomed_image = ndimage.zoom(image, (2, 2, 1))

# Cắt ảnh (crop): Lấy một phần ảnh từ tọa độ (50, 50) đến (150, 150)
cropped_image = image[50:150, 50:150]           

# Thay đổi kích thước ảnh (resize): Giảm kích thước xuống 50%
resized_image = ndimage.zoom(image, (0.5, 0.5, 1))  
```

#### 3. Áp dụng phép biến đổi Affine

```python
# Ma trận Affine cho phép quay và tịnh tiến
affine_matrix = np.array([[1, 0.2, 0], [0.1, 1, 0], [0, 0, 1]])
transformed_image = ndimage.affine_transform(image, affine_matrix)
# Hiển thị các ảnh đã qua xử lý
fig, axs = plt.subplots(1, 4, figsize=(12, 6))
axs[0].imshow(zoomed_image)
axs[0].set_title("Zoomed Image")
axs[1].imshow(cropped_image)
axs[1].set_title("Cropped Image")
axs[2].imshow(resized_image)
axs[2].set_title("Resized Image")
axs[3].imshow(transformed_image)
axs[3].set_title("Affine Transformed Image")
plt.show()
```
