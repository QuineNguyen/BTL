## Chương 5: Image Enhancement Using Derivatives

#### __Bài tập 1:__ Sử dụng hàm unsharp_mask() trong mô-đun skimage.filters với các giá trị khác nhau của tham số radius và amount để làm sắc nét hình ảnh

#### __Giải thích code__

#### 1. Nhập các thư viện cần thiết

```python
# numpy để xử lý mảng
import numpy as np                          
# matplotlib.pylab để vẽ đồ thị và hiển thị ảnh
import matplotlib.pylab as plt              
# imread từ skimage.io để đọc ảnh
from skimage.io import imread               
# unsharp_mask từ skimage.filtes để làm sắc nét ảnh
from skimage.filters import unsharp_mask    
```

#### 2. Đọc ảnh và áp dụng hàm unsharp_mask với các giá trị khác nhau của radius và amount

```python
# Đọc ảnh từ đường dẫn
im = imread('../images/apple.png')          
# Áp dụng hàm unsharp_mask với radius=1 và amount=1 để làm sắc nét ảnh và lưu kết quả vào im1
im1 = unsharp_mask(im, radius=1, amount=1)  
# Áp dụng hàm unsharp_mask với radius=5 và amount=2 để làm sắc nét ảnh và lưu kết quả vào im2
im2 = unsharp_mask(im, radius=5, amount=2)  
# Áp dụng hàm unsharp_mask với radius=20 và amount=3 để làm sắc nét ảnh và lưu kết quả vào im3
im3 = unsharp_mask(im, radius=20, amount=3) 
```

#### 3. Hiển thị và lưu các kết quả

```python
# Tạo một hình ảnh với kích thước 20x12 inch và chia thành 4 ô (2 hàng, 2 cột)
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(20, 12))
axes = axes.ravel()
# Hiển thị ảnh gốc im và các ảnh đã làm sắc nét im1, im2, im3 trong các ô tương ứng
# Đặt tiêu đề cho từng ô với các thông số radius và amount đã sử dụng
axes[0].set_title('Original image', size=20), axes[0].imshow(im)
axes[1].set_title('Enhanced image, radius=1, amount=1.0', size=20), axes[1].imshow(im1)
axes[2].set_title('Enhanced image, radius=5, amount=2.0', size=20), axes[2].imshow(im2)
axes[3].set_title('Enhanced image, radius=20, amount=3.0', size=20), axes[3].imshow(im3)
# Tắt các trục tọa độ cho tất cả các ô
for ax in axes:
    ax.axis('off')
# Sử dụng fig.tight_layout() để tự động điều chỉnh khoảng cách giữa các ô
fig.tight_layout()
# Lưu kết quả thành file unsharp_mask_apple.png
plt.savefig('unsharp_mask_apple.png')
# Hiển thị hình ảnh
plt.show()
```
