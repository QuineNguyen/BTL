## Chương 1: Getting Started with Image Processing

#### __Bài tập 1:__ Sử dụng các hàm của thư viện scikit-image để đọc một tập hợp các hình ảnh và hiển thị chúng dưới dạng một bức ảnh ghép.

##### __Giải thích code__

##### 1. Nhập các thư viện cần thiết

```python
import skimage.io 				# dùng để đọc và lưu ảnh
import skimage.util		        # cung cấp công cụ montage để tạo ảnh ghép từ nhiều ảnh con
```

#### 2. Thực hiện ghép ảnh

```python
# ảnh a được đọc từ tệp images/lena.jpg
a = skimage.io.imread('../images/lena.jpg')	
# các giá trị pixel giảm một nửa
b = a // 2					
# các giá trị pixel chỉ bằng 1/3 so với ảnh gốc                
c = a // 3					                
m = skimage.util.montage([a, b, c], fill=(255, 128, 0), channel_axis=-1)
''' 
skimage.util.montage tạo ra một ảnh ghép từ các ảnh đầu vào [a, b, c]; 
fill=(255, 128, 0) là màu cam nhạt với các giá trị RGB lần lượt là (255, 128, 0); 
channel_axis=-1 chỉ định rằng trục cuối cùng của mảng đại diện cho các kênh màu RGB
'''
# hiển thị ảnh m
skimage.io.imshow(m)  
# lưu lại thành tệp skimage_montage_default.jpg                              
skimage.io.imsave('skimage_montage_default.jpg', m) 
```
